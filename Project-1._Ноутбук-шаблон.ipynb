3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Определяем функцию для преобразования признака "Опыт работы" в месяцы.
def experience_in_months(exp):
    if exp is np.nan or exp == 'Не указано':
        return None
    year_options = ['год', 'года', 'лет']
    month_options = ['месяц', 'месяца', 'месяцев']
    exp = str(exp)
    exp_list = exp.split(' ')[:7]
    year = 0
    month = 0
    for index, value in enumerate(exp_list):
        if value in year_options:
            year = int(exp_list[index-1])
    for index, value in enumerate(exp_list):
        if value in month_options:
            month = int(exp_list[index-1])
    return year*12 + month
    
# Добавляем признак "Опыт работы (месяц)" применив функцию experience_in_months.
hh_resume['Опыт работы (месяц)'] = hh_resume['Опыт работы'].apply(experience_in_months)

# Находим медианное значение опыта работы соискателя в месяцах.
print(hh_resume['Опыт работы (месяц)'].median())

# Удаляем признак "Опыт работы"
hh_resume.drop('Опыт работы', axis=1, inplace=True)

print(hh_resume.info())
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Определяем функцию для выделения признака "Город".
def get_city(arg):
    capital_cities = ['Москва', 'Санкт-Петербург']
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    arg_split = arg.split(' , ')
    city = ''
    if (arg_split[0] in capital_cities):
        city = arg_split[0]
        return city
    elif arg_split[0] in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

# Добавляем признак "Город" с помощью функции get_city.
hh_resume['Город'] = hh_resume['Город, переезд, командировки'].apply(get_city)

print(hh_resume['Город'].value_counts(normalize=True))

# Определяем функцию для выделения признака "Готовность к переезду".
def ready_for_moving(arg):
    ready_to_move_options = ['готов к переезду', 'готова к переезду', 'хочу переехать']
    not_ready_to_move_options = ['не готов к переезду', 'не готова к переезду']
    arg_split = arg.split(' , ')
    for x in not_ready_to_move_options:
        for y in arg_split:
            if y.find(x) == 0:
                return False
    for x in ready_to_move_options:
        for y in arg_split:
            if y.find(x) == 0:
                return True
            
# Добавляем признак "Готовность к переезду" с помощью функции ready_for_moving.
hh_resume['Готовность к переезду'] = hh_resume['Город, переезд, командировки'].apply(ready_for_moving)

print(hh_resume['Готовность к переезду'].value_counts())

# Определяем функцию для выделения признака "Готовность к командировкам".
def ready_for_business_trips(arg):
    ready_for_trips_options = ['готов', 'готова']
    not_ready_for_trips_options = ['не готов', 'не готова']
    arg_split = arg.split(' , ')
    for x in arg_split:
        if 'командировкам' in x:
            for y in not_ready_for_trips_options:
                if x.find(y) == 0:
                    return False
            for y in ready_for_trips_options:
                if x.find(y) == 0:
                    return True
    return False

# Добавляем признак "Готовность к командировкам" с помощью функции ready_for_business_trips.
hh_resume['Готовность к командировкам'] = hh_resume['Город, переезд, командировки'].apply(ready_for_business_trips)

print(hh_resume['Готовность к командировкам'].value_counts())

print(hh_resume[(hh_resume['Готовность к командировкам'] == True) & (hh_resume['Готовность к переезду'] == True)].shape[0]/44744*100)

# Удаляем признак "Город, переезд, командировки"
hh_resume.drop('Город, переезд, командировки', axis=1, inplace=True)


   
            
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Создаем DataFrame с курсам валют
currency_rates = pd.read_csv('C:\SkillFactory\PROJECT_1\ExchangeRates.csv', sep=',')

# Преобразуем признак "Date" в формат даты.
currency_rates['date'] = pd.to_datetime(currency_rates['date'], dayfirst=True)


print(currency_rates[currency_rates['currency'] == 'UZS'].head())

# преобразуем признак "Обновление резюме" в формат даты, оставив только дату и убрав время.
hh_resume['Обновление резюме'] = pd.to_datetime(hh_resume['Обновление резюме'], dayfirst=True)
hh_resume['Обновление резюме'] = hh_resume['Обновление резюме'].dt.date
hh_resume['Обновление резюме'] = pd.to_datetime(hh_resume['Обновление резюме'])

#print(hh_resume['Обновление резюме'].info())

#print(hh_resume['ЗП'][38])

# Добавляем признаки "Сумма ЗП" и "Валюта", выделив эти даные из признака "ЗП".
hh_resume['Сумма ЗП'] = hh_resume['ЗП'].apply(lambda x: float(x.split()[0]))
hh_resume['Наименование валюты в данных'] = hh_resume['ЗП'].apply(lambda x: x.split()[1])

#print(hh_resume['Наименование валюты в данных'].value_counts())

# Создаем DataFrame с сопоставлением наименования валюты в данных официальным названиям.
currencies = pd.DataFrame(data = [['грн.', 'UAH'], ['USD', 'USD'], ['EUR', 'EUR'], ['бел.руб.', 'BYN'], ['KGS', 'KGS'], ['сум', 'UZS'], ['AZN', 'AZN'], ['KZT', 'KZT'], ['руб.', 'RUB']], columns=['Наименование валюты в данных', 'Наименование валюты в ISO-кодировке'])

# Добавляем признак "Наименование валюты в ISO-кодировке" в таблицу с резюме.
hh_resume = hh_resume.merge(currencies, how='outer', on='Наименование валюты в данных')

# добавляем курс валюты
hh_resume = hh_resume.merge(currency_rates, how='left', left_on=['Обновление резюме', 'Наименование валюты в ISO-кодировке'], right_on=['date', 'currency'])

# Заполняем курс валюты и признак 'proportion' для зарплаты в рублях значением 1.
hh_resume['close'] = hh_resume['close'].fillna(value=1)
hh_resume['proportion'] = hh_resume['proportion'].fillna(value=1)

# Добавляем признак "ЗП (руб)".
hh_resume['ЗП (руб)'] = hh_resume['Сумма ЗП'] * hh_resume['close'] / hh_resume['proportion']

# Удаляем лишние признаки.
hh_resume.drop('Сумма ЗП', axis=1, inplace=True)
hh_resume.drop('Наименование валюты в данных', axis=1, inplace=True)
hh_resume.drop('Наименование валюты в ISO-кодировке', axis=1, inplace=True)
hh_resume.drop('currency', axis=1, inplace=True)
hh_resume.drop('per', axis=1, inplace=True)
hh_resume.drop('time', axis=1, inplace=True)
hh_resume.drop('close', axis=1, inplace=True)
hh_resume.drop('date', axis=1, inplace=True)
hh_resume.drop('vol', axis=1, inplace=True)
hh_resume.drop('proportion', axis=1, inplace=True)
hh_resume.drop('ЗП', axis=1, inplace=True)


print(hh_resume.info())
display(hh_resume['ЗП (руб)'].median())
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Построим распределение признака "Возраст" в логарифмическом масштабе.

fig_6, ax_6 = plt.subplots(1, 1, figsize=(8, 4))
log_age = np.log(hh_resume['Возраст'])
histplot = sns.histplot(log_age, bins=30, ax=ax_6)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+ 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('Распределение признака "Возраст" в логарифмическом масштабе');

# Определим функцию для подсчета числа выбросов.

def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score(hh_resume, 'Возраст', log_scale=True)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')

# Печатаем таблицу с выбросами.
display(outliers)

# Удаляем строки с выбросами.
hh_resume = hh_resume.drop(outliers.index, axis=0)

print(f'Результирующее число записей: {cleaned.shape[0]}')
```

Распределение ассиметрично в левую сторону.
В категорию выбросов попали кандидаты с возрастом 15 и 100 лет.
